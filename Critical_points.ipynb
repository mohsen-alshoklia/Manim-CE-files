```python
%%manim -v WARNING --resolution 1080,1920 -qm CriticalPointsScene

from manim import *

class CriticalPointsScene(Scene):
    def construct(self):
        # 1. Title
        title = Text("Critical Points", font_size=80)
        title.set_color_by_gradient(PINK, PURPLE)
        title.to_edge(UP, buff=1.5)
        self.play(Write(title))

        # 2. Axes Setup
        ax = Axes(
            x_range=[-3.5, 6.5, 1],
            y_range=[-30, 70, 10],
            x_length=8,
            y_length=8,
            axis_config={"include_tip": True, "font_size": 24},
            x_axis_config={"numbers_to_include": [-3, -2, -1, 1, 2, 3, 4, 5, 6]},
            y_axis_config={"numbers_to_include": [-20, -10, 10, 20, 30, 40, 50, 60]}
        )

        # Shift axes slightly up to make room for the math at the bottom
        ax.shift(UP * 1.5)

        # Axis labels
        x_label = ax.get_x_axis_label("x")
        y_label = ax.get_y_axis_label("y")
        self.play(Create(ax), Write(x_label), Write(y_label))

        # 3. Functions
        def func(x):
            return x**3 - 3*x**2 - 9*x + 2

        def deriv(x):
            return 3*x**2 - 6*x - 9

        # Draw f(x)
        graph_f = ax.plot(func, color=PINK, x_range=[-2.8, 4.8])
        label_f_graph = MathTex("f(x)", color=PINK).move_to(ax.c2p(3, 30))

        eq_f = MathTex("f(x) = x^3 - 3x^2 - 9x + 2", color=PINK)
        eq_f.next_to(ax, DOWN, buff=1)

        self.play(Create(graph_f), Write(label_f_graph))
        self.play(Write(eq_f))
        self.wait(1)

        # Draw f'(x)
        graph_df = ax.plot(deriv, color=BLUE_C, x_range=[-2.2, 4.2])
        label_df_graph = MathTex("f'(x)", color=BLUE_C).move_to(ax.c2p(2.5, 5))

        eq_df = MathTex("f'(x) = 3x^2 - 6x - 9", color=BLUE_C)
        eq_df.next_to(eq_f, DOWN, buff=0.5)

        self.play(Create(graph_df), Write(label_df_graph))
        self.play(Write(eq_df))
        self.wait(1)

        # 4. Solving for Critical Points
        # Clear original equations to make room for algebra
        self.play(FadeOut(eq_f), eq_df.animate.next_to(ax, DOWN, buff=1))

        eq_zero = MathTex("f'(x) = 0")
        eq_zero.next_to(eq_df, DOWN, buff=0.5)
        self.play(Write(eq_zero))
        self.wait(0.5)

        step1 = MathTex("3x^2 - 6x - 9 = 0").next_to(eq_zero, DOWN, buff=0.4)
        self.play(Write(step1))

        step2 = MathTex("3(x^2 - 2x - 3) = 0").move_to(step1)
        self.play(Transform(step1, step2))
        self.wait(0.5)

        step3 = MathTex("3(x - 3)(x + 1) = 0").move_to(step1)
        self.play(Transform(step1, step3))
        self.wait(0.5)

        roots = MathTex("x = 3, \quad x = -1", color=YELLOW)
        roots.next_to(step1, DOWN, buff=0.5)
        self.play(Write(roots))
        self.wait(1)

        # 5. Visualizing Maxima and Minima on Graph
        # x = -1 (Maxima)
        line_max = ax.get_vertical_line(ax.c2p(-1, func(-1)), color=YELLOW, line_func=DashedLine)
        dot_max = Dot(ax.c2p(-1, func(-1)), color=YELLOW)
        label_max = Text("Maxima", font_size=24, color=PINK).next_to(dot_max, UP)

        # x = 3 (Minima)
        line_min = ax.get_vertical_line(ax.c2p(3, func(3)), color=YELLOW, line_func=DashedLine)
        dot_min = Dot(ax.c2p(3, func(3)), color=YELLOW)
        label_min = Text("Minima", font_size=24, color=PINK).next_to(dot_min, DOWN)

        self.play(Create(line_max), Create(line_min))
        self.play(FadeIn(dot_max, scale=0.5), FadeIn(dot_min, scale=0.5))
        self.play(Write(label_max), Write(label_min))

        # 6. Final Evaluation Text
        self.play(FadeOut(eq_df, eq_zero, step1, roots)) # Clear algebra

        eval_max = MathTex(r"\text{Maxima: } f(-1) = 7", color=PINK)
        eval_min = MathTex(r"\text{Minima: } f(3) = -25", color=PINK)

        eval_group = VGroup(eval_max, eval_min).arrange(DOWN, aligned_edge=LEFT)
        eval_group.next_to(ax, DOWN, buff=1.5)

        self.play(Write(eval_max))
        self.play(Write(eval_min))

        self.wait(2)
```
